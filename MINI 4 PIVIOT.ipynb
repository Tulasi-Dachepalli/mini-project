```python
# Step 1: Get the inputs of birth rate data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python

data = pd.read_csv(r"C:\Users\91939\Downloads\births.csv")
print(data)
```

```python
# Step 2: Create pivot table with index- decade and column 'gender'
data['decade'] = 10 * (data['year'] // 10)
pivot_table = data.pivot_table('births', index='decade', columns='gender', aggfunc='sum')

# Step 3: Plot it with seaborn or Matplotlib
sns.set()
pivot_table.plot()
plt.ylabel('Total births per year')
plt.show()
```
