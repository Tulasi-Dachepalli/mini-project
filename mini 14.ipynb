```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Step 1: Create an empty image
img = np.zeros((786,786,3), np.uint8)

# Step 2: Add text to the image
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img,'Hello cv',(10,500), font, 4,(87,145,212),2,cv2.LINE_4)
cv2.putText(img,'open',(10,435), font, 4,(94,12,224),2,cv2.LINE_8)


# Step 3: Display the image using Matplotlib
plt.imshow(img)
plt.show()
cv2.imwrite('output.png',img)
```
